#Vamos começar **importando** as bibliotecas necessárias e carregando os dados:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Armazenando os dados fornecidos:

```python
data = {
    'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [6, 7, 8, 6, 7, 5, 4],
    'Bugs Corrigidos': [3, 2, 1, 4, 3, 2, 1],
    'Tarefas Concluídas': [5, 4, 6, 4, 5, 3, 2]
}

df = pd.DataFrame(data)
```

# Calcular os:
■ Total de Horas Trabalhadas

■ Média Diária de Horas Trabalhadas

■ Total de Bugs Corrigidos

■ Média Diária de Bugs Corrigidos

■ Total de Tarefas Concluídas

■ Média Diária de Tarefas Concluídas

■ Produtividade Diária (Tarefas Concluídas por Hora)

```python
total_horas_trabalhadas = df['Horas Trabalhadas'].sum()
media_diaria_horas = df['Horas Trabalhadas'].mean()
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()
media_diaria_bugs = df['Bugs Corrigidos'].mean()
total_tarefas_concluidas = df['Tarefas Concluídas'].sum()
media_diaria_tarefas = df['Tarefas Concluídas'].mean()
produtividade_diaria = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
```

# Gráfico de barras para horas trabalhadas por dia

```python
plt.plot(df['Dia'], df['Horas Trabalhadas'], marker='o')
plt.title('Horas Trabalhadas por Dia')
plt.xlabel('Dia da Semana')
plt.ylabel('Horas Trabalhadas')
plt.grid(True)
plt.show()
```

# Gráfico de linha para bugs corrigidos por dia

```python
# Gráfico de linha para bugs corrigidos por dia
plt.figure(figsize=(10, 6))
plt.plot(df['Dia'], df['Bugs Corrigidos'], marker='o', color='green')
plt.title('Bugs Corrigidos por Dia')
plt.xlabel('Dia da semana')
plt.ylabel('Bugs Corrigidos')
plt.grid(True)
plt.show()
```

# Gráfico de linha para tarefas concluídas por dia


```python
plt.figure(figsize=(10, 6))
plt.plot(df['Dia'], df['Tarefas Concluídas'], marker='o', color='orange')
plt.title('Tarefas Concluídas por Dia')
plt.xlabel('Dia da Semana')
plt.ylabel('Tarefas Concluídas')
plt.grid(True)
plt.show()
```

# Apresentação dos resultados e insights em markdown

```python
from IPython.display import display, Markdown

# Relatório de Progresso Diário
display(Markdown(f"""
- **Total de Horas Trabalhadas**: {total_horas_trabalhadas}
- **Média Diária de Horas Trabalhadas**: {media_diaria_horas:.2f}
- **Total de Bugs Corrigidos**: {total_bugs_corrigidos}
- **Média Diária de Bugs Corrigidos**: {media_diaria_bugs:.2f}
- **Total de Tarefas Concluídas**: {total_tarefas_concluidas}
- **Média Diária de Tarefas Concluídas**: {media_diaria_tarefas:.2f}
"""))

# DataFrame com Produtividade Diária
display(Markdown("## DataFrame com Produtividade Diária:"))
display(df)
```

#DashBoard Finale! ⛹

```python
# Redefinir a ordem dos dias
ordered_days = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']
df = pd.DataFrame(data)

# Adicione esta linha aqui
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']

df['Dia'] = pd.Categorical(df['Dia'], categories=ordered_days, ordered=True)
df = df.sort_values('Dia')


# Configurar o estilo dos gráficos
sns.set(style="whitegrid")

# Criar a figura e os eixos para o dashboard
fig = plt.figure(figsize=(20, 8))  # Largura ajustada para 15
ax1 = plt.subplot2grid((2, 3), (0, 0), colspan=3)
ax2 = plt.subplot2grid((2, 3), (1, 0))
ax3 = plt.subplot2grid((2, 3), (1, 1))
ax4 = plt.subplot2grid((2, 3), (1, 2))

# Gráfico de Total de Horas Trabalhadas
ax1.plot(df['Dia'], df['Horas Trabalhadas'], marker='o', color='skyblue')
ax1.set_title('Horas Trabalhadas por Dia')
ax1.set_ylabel('Horas Trabalhadas')
ax1.set_xlabel('')

# Gráfico de Produtividade Diária

sns.barplot(x='Dia', y='Produtividade Diária', data=df, ax=ax2, color='lightgreen')

ax2.set_title('Produtividade Diária')
ax2.set_ylabel('Produtividade (Tarefas/Hora)')
ax2.set_xlabel('')
ax2.grid(True)

# Gráfico de Total de Tarefas Concluídas
sns.barplot(x='Dia', y='Tarefas Concluídas', data=df, ax=ax3, color='salmon')
ax3.set_title('Tarefas Concluídas por Dia')
ax3.set_ylabel('Tarefas Concluídas')
ax3.set_xlabel('')
ax3.grid(True)

# Gráfico de Total de Bugs Corrigidos
sns.barplot(x='Dia', y='Bugs Corrigidos', data=df, ax=ax4, color='skyblue')
ax4.set_title('Bugs Corrigidos por Dia')
ax4.set_ylabel('Bugs Corrigidos')
ax4.set_xlabel('')
ax4.grid(True)

# Criar a tabela de métricas
fig, ax5 = plt.subplots(figsize=(15, 3)) # Ajustar o tamanho da figura para a tabela
ax5.axis('off')
table_data = [
    ["Total de Horas Trabalhadas", total_horas_trabalhadas],
    ["Média Diária de Horas Trabalhadas", f"{media_diaria_horas:.2f}"],
    ["Total de Bugs Corrigidos", total_bugs_corrigidos],
    ["Média Diária de Bugs Corrigidos", f"{media_diaria_bugs:.2f}"],
    ["Total de Tarefas Concluídas", total_tarefas_concluidas],
    ["Média Diária de Tarefas Concluídas", f"{media_diaria_tarefas:.2f}"]

]
table = ax5.table(cellText=table_data, colLabels=["Métricas", "Valores"], cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(14)
table.scale(1.2, 1.2)

plt.tight_layout()

plt.show()
```
